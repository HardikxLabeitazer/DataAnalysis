```python
def get_row_count(*dfs):
  return [df.shape[0] for df in dfs]

get_row_count(station_info,weather)
```

```python
inner_join = weather.merge(
    station_info,left_on = 'station',right_on='id'
)

inner_join.sample(5,random_state=0)
```

```python
weather.merge(
    station_info.rename(dict(id='station'),axis=1),
    on='station'
).sample(5,random_state=0)
```

```python
left_join = station_info.merge(
    weather,left_on= 'id',right_on='station',how='left'
)
right_join = weather.merge(
    station_info,left_on = 'station',right_on='id',how='right'
)
right_join[right_join.datatype.isna()].head()
```

```python
left_join.sort_index(axis=1).sort_values(['date','station'],ignore_index=True).equals(right_join.sort_index(axis=1).sort_values(['date','station'],ignore_index=True))
```

```python
get_row_count(inner_join,left_join,right_join)
```

```python
outer_join = weather.merge(
    station_info[station_info.id.str.contains('US1NY')],
    left_on ='station',right_on='id',
    how='outer',indicator=True
)
```

```python
pd.concat([
           outer_join.query(f'_merge =="{kind}"').sample(2,random_state=0) for kind in outer_join._merge.unique()
]).sort_index()
```

```python
dirty_data = pd.read_csv('/content/drive/MyDrive/DataAnalysis/Ch4/dirty_data.csv',
                         index_col='date').drop_duplicates().drop(columns='SNWD')
dirty_data.head()
```

```python
valid_station = dirty_data.query('station != "?" ').drop(columns =['WESF','station'])
```

```python
station_with_wesf = dirty_data.query('station=="?"').drop(columns=['station','TOBS','TMIN','TMAX'])
```

```python
valid_station.merge(
    station_with_wesf,how='left',
    left_index=True,right_index=True
).query('WESF>0').head()
```

```python
valid_station.merge(
    station_with_wesf,how='left',
    left_index=True,right_index=True,
    suffixes=('','_?')
).query('WESF>0').head()
```

```python
valid_station.join(
    station_with_wesf,how='left',rsuffix='_?'
).query('WESF>0').head()
```
